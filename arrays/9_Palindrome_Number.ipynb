# 9. Palindrome Number

<h4>Problem Description:</h4>
Given an integer x, return true if x is a Palindrome and false otherwise.
<b>Example 1:</b>
Input: x = 121
Output: true
Explanation: 121 reads as 121 from left to right and from right to left.

<b>Example 2:</b>
Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

<b>Example 3:</b>
Input: x = 10
Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.



<h4>Solution 1: Time complexity: 3ms Memory Space: 17.84MB Complexity: $$O(n)$$</h4>

```python
class isPalindrome:
    def check(self, number):
        conversion_string = str(number)
        reversed_number = conversion_string[::-1]
        if reversed_number == conversion_string:
            return True
        else:
            return False
```

```python
test = isPalindrome()
print(test.check("121"))
```

<h4>Solution 2: Time complexity: 22ms Memory Space: 17.59MB Complexity: $$O(n)$$</h4>

```python
class isPalindrome:
    def check_2(self, number):
        temp = number
        reverse = ""
        while number > 0:
            rem = number % 10
            reverse = reverse + str(rem)
            number = number // 10
            
        if reverse == str(temp):
            return True
            
        elif temp == 0:
            return True

        else:
            return False
```

```python
test_2 = isPalindrome()
print(test_2.check_2(0))
```

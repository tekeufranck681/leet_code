# 1. TWO SUM 

<b>Problem Description</b>

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
You may assume that each input would have exactly one solution, and you may not use the same element twice.
You can return the answer in any order.

Example 1:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:
Input: nums = [3,2,4], target = 6
Output: [1,2]

Example 3:
Input: nums = [3,3], target = 6
Output: [0,1]


<h3>Solutions</h3>

<h4>First Try. Runtime: <b>1869ms</b> Memory:<b>18.28MB</b> Complexity: <b>O(n^2)</b></h4>

```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(len(nums)-1):
            for j in range(i+1, len(nums)):
                total = nums[i] + nums[j]
                if total == target:
                    return [i, j]
test = Solution()
print(test.twoSum([2,7,11,15], 9))
```

<h4>Second Try. Runtime: <b>219ms</b> Memory:<b>18.5MB</b> Complexity: <b>O(n^2)</b></h4>

```python
class Solution:
    def twoSum(self, nums, target):
        count = 0
        n = len(nums)
        for i in range(n):
            rem = target - nums.pop(0)
            count = count + 1
            if rem in nums:
                return [i, nums.index(rem) + count ]
        return []
test2 = Solution()
print(test2.twoSum([3,2,4], 6))
```

<h4>Third Try: Runtime: <b>0ms</b> Memory:<b>19.10MB</b> Complexity: <b>O(n)</b></h4>

```python
class Solution:
    def twoSum(self, nums, target):
        seen = {}
        for i, num in enumerate(nums):
            diff = target - num
            if diff in seen:
                return [seen[diff], i]
            seen[num] = i
        return []
test3 = Solution()
print(test3.twoSum([3,3],6))
```
